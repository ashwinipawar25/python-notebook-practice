# Bar graph

```python
x = [1,2,3,4]
y = [5,6,7,8]
plt.bar(x,y)
plt.show()
```

```python
x = [1,2,3,4]
y = [5,6,7,8]
c =['r','green','orange','blue']
plt.bar(x,y,color = c,width = 0.5)
plt.show()
```

```python
import numpy as np
x = np.array(['A','B','C','D'])
y = np.array([3,8,1,10])
cd = ['r','black','orange','blue']
plt.barh(x,y,color = 'hotpink')
plt.show()
```

# Scatter plot

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [6,7,8,9,10]
plt.title("Scatter plot")
plt.scatter(x,y,marker = '*')
plt.xlabel("month")
plt.ylabel("number")
plt.show()
```

# pie graph

```python
x = [10,20,30,40,50]
y = ['sql',"python",'excel','power bi','tableu']

plt.pie(x,labels = y)
plt.show()
```

```python
x = [10,10,15,25,40]
y = ['statistics','hindi','english','maths','science']
c  = ['red','blue','orange','yellow','aqua']

plt.pie(x ,labels = y,colors = c)
plt.legend()
plt.show()
```
